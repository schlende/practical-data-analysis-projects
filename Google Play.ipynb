```python
import urllib.robotparser

rp = urllib.robotparser.RobotFileParser()
rp.set_url("https://www.etsy.com/robots.txt")
rp.read()
rp.crawl_delay("*")

test_crawl_url="https://play.google.com/store/apps/details?id=com.wildnotion.poetscorner"

can_crawl_listings = rp.can_fetch("*", test_crawl_url)
print("We can crawl Google Play? {0}".format(can_crawl_listings)) 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
df = pd.read_csv(r"C:\Users\gosharm\Downloads\TEST\Data Science\practical-pandas-projects_Gourav\datasets\google-play-store-11-2018.csv", index_col= "app_id")
```

```python
from datetime import date
import datetime as dt
today = date.today()
df["Current Day"] = today
df["No_Of_Days"] = (pd.to_datetime(df["Current Day"]) - pd.to_datetime(df["released"])).dt.days
df.rename(columns={'rating_one_star': '1 Star', 'rating_two_star': '2 Star', 'rating_three_star': '3 Star', 'rating_four_star': '4 Star', 'rating_five_star': '5 Star'}, inplace=True)
df['Avg_download'] = round(df['min_installs']/df['No_Of_Days'])
#df.set_index("app_id", inplace=True)
df.columns = [x.title() for x in df.columns]
df.sort_index()
```
